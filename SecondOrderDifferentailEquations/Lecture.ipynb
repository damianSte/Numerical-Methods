```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dx_dt(x,v):
    return v
```

```python
def dv_dt(x,v):
    return -x
```

```python
x0 = 1
v0 = 0
n = 100
x_values =[]
v_values = []
t_values = []
energy_values = []
dt = 2*np.pi/n
t = 0

for i in range(n):
    x0 += dx_dt(x0, v0)*dt
    v0 += dv_dt(x0, v0)*dt
    t += dt

    x_values.append(x0)
    v_values.append(v0)
    energy_values.append(0.5*v0**2 + 0.5*x0**2)
    t_values.append(t)
```

```python
plt.plot(x_values, v_values)
```

```python
plt.plot(t_values, energy_values)
```
