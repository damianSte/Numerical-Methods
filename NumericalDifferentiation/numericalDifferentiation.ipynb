```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

Ex.1 

- We need to begin with calculating derivative of a given polynomial function f(x)

```python
def f(x):
    return -0.1 * x**4 - 0.15 * x**3 - 0.5 * x**2 - 0.25 * x + 1.2

def f_prime(x):
    return -0.4 * x ** 3 - 0.45 * x ** 2 - 1.0 * x - 0.25
```

- We calculate value of the function f'(x) when x = .5

```python
x = 0.5
f_prime_val = f_prime(x)
f_prime_val
```

- We compute the difference quotient for a chosen, finite h
- h = 10^-n 
- we calculate absolute error for f'(x) - f(a)
- we repeat calculations (using loop) 11 times

```python
n_values  = np.arange(1,12 ,dtype=float)
h_values = 10 **(-n_values)

f_prime_numerical = (f(x + h_values) - f(x)) / h_values

absolute_errors = np.abs(f_prime_numerical - f_prime_val)

for i in range(len(n_values)):
    print(n_values[i], absolute_errors[i])
```

```python
plt.figure(figsize=(10, 6))
plt.loglog(n_values, absolute_errors, marker='o', color='orange')
plt.title("Absolute Errors of Numerical Derivative")
plt.xlabel("n (where h = 10^(-n))")
plt.ylabel("Absolute Error")
plt.grid(True)
plt.show()
```

```python
def f_exp_prime(x):
    return  np.exp(np.sin(2 * x)) * 2 * np.cos(2 * x)
```

```python
x = 0.5
f_prime_val_exp = f_exp_prime(x)
print(f_prime_val_exp)
```

```python
n_values_sine = np.arange(1, 12, dtype=float)
h_values_sine = 10 ** (-n_values_sine)
```

```python
f_exp_prime_numerical = (f_exp(x + h_values_sine) - f_exp(x)) / h_values_sine
absolute_errors_exp = np.abs(f_exp_prime_numerical - f_prime_val_exp)

for i in range(len(n_values_sine)):
    print(n_values_sine[i], absolute_errors_exp[i])
```

```python
plt.figure(figsize=(10, 6))
plt.loglog(n_values_sine, absolute_errors_exp, marker='o', color='orange')
plt.title(r"Absolute Errors of Numerical Derivative of e^sin(2x)")
plt.xlabel("n (where h = 10^(-n))")
plt.ylabel("Absolute Error")
plt.grid(True)
plt.show()
```
