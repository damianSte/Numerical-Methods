```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**3 - 2*x**2 + 3 * x -1

def f_prime(x):
    return 3 * x**2 - 4 * x + 3

def f_second_derivative(x):
    return 6 * x - 4
```

```python
f_num = f(0.5)
f_prime_num = f_prime(0.5)
f_second_derivative_num = f_second_derivative(0.5)
```

```python
x = 0.5
n_values = np.arange(1, 12, dtype=float)
h_values = 10 ** (-n_values)
```

```python
f_2nd_dev_numerical = (4 * (f(x + (h_values / 2)) - 2 * f(x) + f(x - (h_values / 2)))) / np.power(h_values, 2)
```

```python
absolute_errors_2nd_dev = np.abs(f_2nd_dev_numerical - f_second_derivative_num)

for i in range(len(n_values)):
    print(n_values[i], absolute_errors_2nd_dev[i])
```

```python
plt.figure(figsize=(10, 6))
plt.loglog(h_values, absolute_errors_2nd_dev, marker='o', color='orange')
plt.title(r"Absolute Errors of Numerical Derivative of e^sin(2x)")
plt.xlabel(" h = 10^-n")
plt.ylabel("Absolute Error")
plt.grid(True)
plt.show()
```

**Conclusions:**

The absolute error of the numerical derivative of the second derivative of the function f(x) = x^3 - 2x^2 + 3x - 1 is decreasing as the step size h is decreasing. Which is bazaar as in previously tested function we saw that the error was increasing as the step size was decreasing. This is due to the fact that the function f(x) = x^3 - 2x^2 + 3x - 1 is a polynomial function and the numerical derivative of the second derivative of a polynomial function is linear?. And we do not get optimal h value.
